```python
# create a list 'null' to contain the columns which have more than 50% missing values
null = [f for f in df.columns if df[f].isnull().sum() > df.shape[0]/2]
null
```

```python
# drop the columns which have more missing values than half the number of records along with the 'Id' column
df.drop(columns = null + ['Id'], inplace = True, axis = 1)
```

```python
# create a function to impute the missing values in the numerical columns with the mean and the categorical features with
# the mode

def fill_missing_values(df, features, method='mean'):
     for feature in features:
        if method == 'mean':
            fill_value = df[feature].mean()
        elif method == 'mode':
            fill_value = df[feature].mode().iloc[0]
        
        df[feature].fillna(value=fill_value, inplace=True)
        
     return df
```

```python
# create a list of numerical_feature

numeric_features = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
null_numeric_features = [col for col in numeric_features if df[col].isnull().any()]
null_numeric_features
```

```python
# view the numerical features with null values and the number of null values in them

for feature in numeric_features:
    num_null = df[feature].isnull().sum()
    if num_null > 0:
        print(f"The feature '{feature}' has {num_null} missing values.")
```

```python
# fill the missing values in 'LotFrontage' with the mean
fill_missing_values(df, ['LotFrontage'])

# i didn't fill 'MasVnrArea' with mean because there are only few missing values and most of the values are 0
# even if 'GarageYrBlt' is a numerical_feature we don't want to fill the missing values with the mean!
```

```python
# make a list of categorical variables
categorical_features = [col for col in df.columns if df[col].dtype == 'object']

# use the above declard function to fill the missing values in 'MasVnrArea' and 'GarageYrBlt' with the mode
fill_missing_values(df, categorical_features + ['MasVnrArea', 'GarageYrBlt'], method = 'mode')
```

```python
# view the missing values using heatmap
sns.heatmap(df.isnull(), yticklabels = False, cbar = False)

# there are no more missing values
```

```python
# viewing the number of unique features in each categorical variable to comapre them with the number of unique features 
# in the test set

for feature in categorical_features:
    print(feature, df[feature].nunique())
```

```python
# since the number of unique features in some categorical variables are different in the train set and the test set, i'm
# combining the dataframes to encode the categarical variables

main_df = df.copy()
test_df = pd.read_csv('formulatedtest.csv')
final_df = pd.concat([df, test_df], axis = 0)
```

```python
# the test dataframe has one less column since it does not have the targer column
test_df.shape
```

```python
# view the shape of the cancatenated dataframe
final_df.shape
```

```python
# function to one hot encode the categorical variables

def category_onehot_multcols(multcolumns):
    df_final = final_df
    for field in multcolumns:
        df1 = pd.get_dummies(final_df[field], drop_first=True, prefix=field)
        df_final = pd.concat([df_final, df1], axis=1)
        df_final.drop(field, axis=1, inplace=True)
    return df_final
```

```python
# passing the categorical columns into the above defined function
final_df = category_onehot_multcols(categorical_features)
```

```python
# the shape of the dataframe after encoding the categoricalfeatures
final_df.shape
```

```python
# seperate out the combined test and train dataframe

df_train = final_df.iloc[:1460]
df_test = final_df.iloc[1460:]
```

```python
df_test.head()
```

```python
# drop the target variable from the test dataset since it was only added when we combined it with train dataset
df_test.drop(['SalePrice'], axis = 1, inplace = True)
```

```python
df_test.shape
```

```python
# divide the training dataframe to seperate the predictor variables and target variable

X_train = df_train.drop(['SalePrice'], axis = 1)
y_train = df_train['SalePrice']
```

```python
import xgboost

# create an instance of xgboost regressor
classifier =  xgboost.XGBRegressor()

# train the model
classifier.fit(X_train, y_train)
```

```python
# predict the output for the test dataframe
y_pred = classifier.predict(df_test)
```

```python
y_pred
```

```python
# since I want to upload the predicted values to Kaggle Competition, I'm following these steps:

# create a dataframe of the predicted values
pred = pd.DataFrame(y_pred)

# read the submission template (csv file) provided by Kaggle
sub_df = pd.read_csv('sample_submission.csv')

# combine the 'Id' column from sample csv file with the dataframe of the predicted values
combined_df = pd.concat ([sub_df['Id'],pred], axis=1)

# assign columns names to the concatenated dataframe
combined_df.columns = ['Id', 'SalePrice']

# create a new csv file of the combined dataframe for submission in Kaggle website
combined_df.to_csv('sample_submission_xgb1.csv', index=False)

# for the above model, I got a score of 0.14121 on Kaggle
```

```python
from sklearn.model_selection import RandomizedSearchCV

# hyperparamater tuning with number of iterations for the randomized search set to 50

# Create an instance of the XGBRegressor class
regressor=xgboost.XGBRegressor()

# Define the hyperparameters to search over
booster=['gbtree','gblinear']
base_score=[0.25,0.5,0.75,1]
n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
booster=['gbtree','gblinear']
learning_rate=[0.05,0.1,0.15,0.20]
min_child_weight=[1,2,3,4]

# Define the grid of hyperparameters to search over
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'learning_rate':learning_rate,
    'min_child_weight':min_child_weight,
    'booster':booster,
    'base_score':base_score
    }

# Create a RandomizedSearchCV object with the defined hyperparameter grid and other parameters
random_cv = RandomizedSearchCV(estimator=regressor,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=42)

# Fit the RandomizedSearchCV object on the training data to find the best hyperparameters
random_cv.fit(X_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
random_cv.best_score_
```

```python
best_estimator = random_cv.best_estimator_

# train the model using the best estimator that we get from the above grid
best_estimator.fit(X_train, y_train)

# Make predictions on the test data
y_pred = best_estimator.predict(df_test)
```

```python
# since I want to upload the predicted values to Kaggle Competition, I'm following these steps:

# create a dataframe of the predicted values
pred = pd.DataFrame(y_pred)

# read the submission template (csv file) provided by Kaggle
sub_df = pd.read_csv('sample_submission.csv')

# combine the 'Id' column from sample csv file with the dataframe of the predicted values
combined_df = pd.concat ([sub_df['Id'],pred], axis=1)

# assign columns names to the concatenated dataframe
combined_df.columns = ['Id', 'SalePrice']

# create a new csv file of the combined dataframe for submission in Kaggle website
combined_df.to_csv('sample_submission_xgb2.csv', index=False)

# for the above model, I got a score of 0.13421 on Kaggle
```
